```python
import pandas as pd
```

```python
df2015 = pd.read_csv('2015.csv', encoding='euc-kr', index_col='관서명')
df2016 = pd.read_csv('2016.csv', encoding='euc-kr', index_col='관서명')
df2017 = pd.read_csv('2017.csv', encoding='euc-kr', index_col='관서명')
```

```python
df2015['총계'] = df2015.loc[:, '살인':'폭력'].sum(axis = 1)
df2016['총계'] = df2016.loc[:, '살인':'폭력'].sum(axis = 1)
df2017['총계'] = df2017.loc[:, '살인':'폭력'].sum(axis = 1)
```

```python
df2017
```

```python
d2015_sum = df2015[df2015['구분'] == '발생건수'].loc[:,'총계']
d2016_sum = df2016[df2016['구분'] == '발생건수'].loc[:,'총계']

d2017 = df2017[df2017['구분'] == '발생건수']
d2017_sum = d2017[d2017.index != '광주지방경찰청'].loc[:, '총계']

d15_16 = (d2016_sum-d2015_sum)/d2015_sum*100
d16_17 = (d2017_sum-d2016_sum)/d2016_sum*100

d2015_sum.name = '2015년총계'
d2016_sum.name = '2016년총계'
d2017_sum.name = '2017년총계'
d15_16.name = '2015~2016증감률'
d16_17.name = '2016~2017증감률'
```

```python
pd.concat([d2015_sum, d15_16, d2016_sum, d16_17, d2017_sum], axis = 1)
```
